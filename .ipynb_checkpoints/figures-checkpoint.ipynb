# 散点图

```python
import matplotlib.pyplot as plt
import numpy as np
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X) # for color value, just for good looking

# c：color
# cmap：这里用的默认
plt.scatter(X,Y,s=75,c=T,alpha=0.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

# 坐标值都弄没有
plt.xticks(())
plt.yticks(())

plt.show()
```

# 柱状图

```python
import matplotlib.pyplot as plt
import numpy as np

n = 12
X = np.arange(n)
Y1 = np.random.uniform(0.5,1.,n) / float(n)
Y2 = np.random.uniform(0.5,1.,n) / float(n)

# 柱状图
plt.bar(X,+Y1,facecolor='#eeeeee',edgecolor='#ffffff')
plt.bar(X,-Y2,facecolor='#9999ff',edgecolor='#ffffff')

# 值写在上面
for x,y in zip(X,Y1):
    # ha : horizontal alignment : 怎样对齐
    plt.text(x,y+0.01,'%.2f'%y,ha='center',va='bottom')

for x,y in zip(X,Y2):
    # ha : horizontal alignment : 怎样对齐
    plt.text(x,-y-0.01,'%.2f'%y,ha='center',va='top')


# 坐标值都弄没有
plt.ylim(-0.15,0.15)
plt.xticks(())
plt.yticks(())

plt.show()
```

# 等高线

```python
import matplotlib.pyplot as plt
import numpy as np

# 用x和y计算对应高度
def f(x,y):
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2)

n = 256
# linspace(min,max,点的数量)
# arange(min,max,每两点之间的间隔)
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)

# 定义网格
X,Y = np.meshgrid(x,y)

# 颜色
# cool/hot都可以
# 10:在高度上分成10份
plt.contourf(X,Y,f(X,Y),10,alpha=0.75,cmap='cool')

# 画等高线
C = plt.contour(X,Y,f(X,Y),10,colors='black',linewidth=0.5)

# 等高线的数字描述
plt.clabel(C,inline=True,fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```

# 图像

```python
import matplotlib.pyplot as plt
import numpy as np

# image data
a = np.random.uniform(0,1,9).reshape(3,3)

# origin:cmap的方向正反
# interpolation：nearest边框分明 / bicubic边框模糊，还有其他很多
plt.imshow(a,interpolation='bicubic',cmap='hot',origin='lower')

# 添加colorbar（也就是cmap）
# shrink压缩高度，原来的0.8
plt.colorbar(shrink=0.8)

plt.xticks(())
plt.yticks(())
plt.show()
```

# 3D 数据

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-3,3,0.1)
Y = np.arange(-3,3,0.1)
X,Y = np.meshgrid(X,Y)

# 用x和y计算对应高度
def f(x,y):
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2)

# height value
Z = f(X,Y)

# 画在3d坐标轴上
# cstride 列跨度
# rstride 行跨度
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='rainbow')

# 增加等高线
# zdir：沿着哪个方向压
# offset：压到的图片放到哪个坐标上
ax.contourf(X,Y,Z,zdir='z',offset=-1,cmap='rainbow')
# 因为上面压到了-1，所以这里的z从-1开始
ax.set_zlim(-1,1)


plt.show()
```

# subplot其他应用
### 1.subplot2grid

```python
import matplotlib.pyplot as plt

plt.figure()
# 3行3列，从0行0列开始，跨3列，跨1行
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)
ax1.plot([1,2,3],[1,2,5])
ax1.set_xlabel("ax1's x label")
ax1.set_ylabel("ax1's y label")
ax1.set_title('ax1 title')
ax1.set_xticks(())
ax1.set_yticks(())


ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)
ax2.plot([1,2,3],[1,2,-7])

ax3 = plt.subplot2grid((3,3),(2,0),colspan=1,rowspan=1)
ax3.plot([1,2,3],[1,2,-7])

ax4 = plt.subplot2grid((3,3),(2,1),colspan=1,rowspan=1)
ax4.plot([1,2,3],[1,2,-7])

ax5 = plt.subplot2grid((3,3),(1,2),colspan=1,rowspan=2)
ax5.plot([1,2,3],[1,2,-7])


plt.plot()
plt.show()
```

### 2.gridspec

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure(figsize=(8,8))
# 有3行3列
gs = gridspec.GridSpec(3,3)
# 第0行，所有列
ax1 = plt.subplot(gs[0,:])

# 第1行，前2列
ax2 = plt.subplot(gs[1,:2])

# 第1，2行，第3列
ax3 = plt.subplot(gs[1:,2])


ax4 = plt.subplot(gs[2,0])
ax5 = plt.subplot(gs[2,1])


plt.show()
```

# 图中图

```python
import matplotlib.pyplot as plt

fig = plt.figure()
x=[1,2,3,4,5,6,7]
y=[1,2,4,3,7,6,7]

# 这里的0.1是相对于整个figure的10%的意思
left,bottom,width,height=0.1,0.1,0.8,0.8
ax1 = fig.add_axes([left,bottom,width,height])
ax1.plot(x,y,'r')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('big fig')


# 从左到右20%，从下到上60%
left,bottom,width,height=0.2,0.6,0.25,0.25
ax2 = fig.add_axes([left,bottom,width,height])
ax2.plot(x,y,'b')
ax2.set_title('small fig')



plt.show()
```

# 次坐标轴（x轴共享，y轴各异）

```python
import matplotlib.pyplot as plt
import numpy as np

# 造数据
x = np.arange(0,10,0.1)
y1 = x**2 - 1
y2 = -y1


fig,ax1 = plt.subplots()
# 给x轴一个镜像，意思是共用一个x轴，各用一个y轴
ax2 = ax1.twinx()
ax1.plot(x,y1,'b-')
ax1.set_xlabel('x label')
ax1.set_ylabel('y1 data',color='b')

ax2.plot(x,y2,'g:')
ax2.set_xlabel('x label')
ax2.set_ylabel('y2 data',color='g')

plt.plot()
plt.show()
```

# 动画(jupyter 插件不行，放到了同目录下animation.py，代码完全相同)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

fig,ax = plt.subplots()

x = np.arange(0,2*np.pi,0.01)
line, = ax.plot(x,np.sin(x))

# animation function传入的值是当前frame的number
def animate(i):
    # 更新ydata的值
    line.set_ydata(np.sin(x+i/100))
    return line,

def init_animate():
    line.set_ydata(np.sin(x))
    return line,



# func:每一帧更新的函数
# frames:100个update
# init_func：最开始的一帧
# interval:多少毫秒更新一次
# blit：是否局部更新（局部更高效，但mac只能选false）
ani = animation.FuncAnimation(fig=fig,func=animate,frames=100,init_func=init_animate,interval=20,blit=False)

# plt.plot()
plt.show()
```
